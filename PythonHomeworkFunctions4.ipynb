```python
# Обозначение глобальных переменных из условия
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

```python
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def handle_command_s(dirs):
    '''
    Обрабатывает команду 's': спрашивает номер документа и выводит полку
    '''
    number = input("\nВведите номер документа:\n").strip()
    shelf = get_shelf_by_number(number, dirs)
    if shelf:
        print(f"\nДокумент хранится на полке: {shelf}\n")
    else:
        print("\nДокумент не найден на полках\n")
```

```python
def get_owner_by_number(doc_number: str, docs):
    '''
    Возвращает имя владельца по номеру документа
    Если документ не найден, то возвращает None
    '''
    doc_number = doc_number.strip()
    for doc in docs:
        if doc.get('number') == doc_number:
            return doc.get('name')
    return None
```

```python
def get_shelf_by_number(doc_number: str, dirs):
    '''
    Возвращает номер полки (ключ словаря directories) на которой лежит документ
    Если документ не найден, то возвращает None
    '''
    doc_number = doc_number.strip()
    for shelf, doc_list in dirs.items():
        if doc_number in doc_list:
            return shelf
    return None
```

```python
def handle_command_p(docs):
    '''
    Обрабатывает команду 'p': спрашивает номер документа и выводит владельца.
    '''
    number = input("\nВведите номер документа:\n").strip()
    owner = get_owner_by_number(number, docs)
    if owner:
        print(f"\nВладелец документа: {owner}\n")
    else:
        print("\nВладелец документа: владелец не найден\n")
```

```python
def main(docs, dirs):
    '''
    Основной цикл программы. Принимает команды от пользователя:
     p: найти владельца документа по номеру
    s:найти полку по номеру документа
      q: выход
    Цикл продолжается, пока пользователь не введёт 'q'
    '''
    prompt = (
        "Введите команду:\n"
    )
    while True:
        cmd = input(prompt).strip().lower()
        if cmd == 'q':
            print("Конец программы!")
            break
        elif cmd == 'p':
            handle_command_p(docs)
        elif cmd == 's':
            handle_command_s(dirs)
        else:
            print("Неизвестная команда. Попробуйте снова.")
```

```python
main(documents, directories)
```
